```python
import numpy # numbers
import scipy.special # logit function
import matplotlib.pyplot as plt # plotting
%matplotlib inline
```

```python
data = open("NNCar.csv")
dataList = data.readlines()
data.close()
```

```python
x = 1
f, axarr = plt.subplots(2,5)
for a in range(0,2):
    for b in range(0,5):
        allValues = dataList[x].split(',')
        image_array = numpy.asfarray(allValues[1:]).reshape((48,48))
        axarr[a,b].imshow(image_array, cmap='Greys', interpolation = 'None')
        x = x + 1
```
